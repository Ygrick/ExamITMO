```python
def standardize_to_usd(
    df_transactions: pd.DataFrame,
    df_exchange: pd.DataFrame
) -> pd.DataFrame:
    """
    Добавляет в df_transactions колонки с эквивалентом в USD:
    - amount_usd,
    - total_amount_usd,
    - max_single_amount_usd.

    Параметры
    ----------
    df_transactions : pd.DataFrame
        Строки транзакций с полями:
        ['date', 'amount', 'currency', 'total_amount', 'max_single_amount'].
    df_exchange : pd.DataFrame
        Курсы валют по датам:
        столбец 'date' + колонки по коду валют (значение — единиц валюты за 1 USD).

    Возвращает
    -------
    pd.DataFrame
        Исходный df_transactions с доп. колонками:
        ['amount_usd', 'total_amount_usd', 'max_single_amount_usd'].
    """
    # Копируем и приводим даты
    df_transactions['date'] = pd.to_datetime(df_transactions['timestamp'].dt.date)
    df_exchange['date'] = pd.to_datetime(df_exchange['date'])

    # Переводим df_exchange в «длинный» формат: date | currency | rate
    exchange_long = (
        df_exchange
        .melt(id_vars=['date'], var_name='currency', value_name='rate')
    )

    # Объединяем по date + currency
    df_merged = df_transactions.merge(exchange_long, on=['date', 'currency'], how='left')

    # Конвертация: rate — единиц валюты за 1 USD
    df_merged['amount'] = df_merged['amount'] / df_merged['rate']
    df_merged['total_amount'] = df_merged['total_amount'] / df_merged['rate']
    df_merged['max_single_amount'] = df_merged['max_single_amount'] / df_merged['rate']
    df_merged = df_merged.drop(columns=['date', 'rate', 'currency'])
    df_merged.to_parquet('transactions_usd.parquet')
    return df_merged

standardize_to_usd(df_transactions, df_exchange)
```

### Мы стандартизировали валюты в USD, теперь нам не нужно хранить таблицу с курсами валют, чтоб снижает нагрузка на память.

```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
df_transactions = pd.read_parquet('transactions_usd.parquet')

# Сразу выведем много информации, чтоб сюда постоянно поглядывать
print('Посмотрим на первые строки:')
display(df_transactions.head())

print('Посмотрим на информацию о данных:')
display(df_transactions.info())

print('Посмотрим на описательные статистики:')
display(df_transactions.describe())

print('Посмотрим на количество nulls:')
display(df_transactions.isnull().sum())

print('Посмотрим на количество дубликатов:')
display(df_transactions.duplicated().sum())

print('Посмотрим на количество уникальных значений:')
display(df_transactions.nunique())

print('Посмотрим на количество строк и столбцов:')
display(df_transactions.shape)

print('Посмотрим на названия столбцов:')
display(df_transactions.columns)
```

```python
# Анализ типов данных для оптимизации памяти
print("=== АНАЛИЗ ТИПОВ ДАННЫХ ДЛЯ ОПТИМИЗАЦИИ ===")
print(f"Общий размер в памяти: {df_transactions.memory_usage(deep=True).sum() / 1024**3:.2f} GB")
print()

# Анализируем каждый столбец
for col in df_transactions.columns:
    dtype = df_transactions[col].dtype
    memory_mb = df_transactions[col].memory_usage(deep=True) / 1024**2
    
    print(f"{col:25} | {str(dtype):15} | {memory_mb:6.1f} MB", end="")
    
    if dtype == 'object':
        unique_count = df_transactions[col].nunique()
        total_count = len(df_transactions)
        ratio = unique_count / total_count
        print(f" | Уникальных: {unique_count:7} ({ratio:.4f})")
    elif dtype in ['int64', 'Int64']:
        min_val = df_transactions[col].min()
        max_val = df_transactions[col].max()
        print(f" | Диапазон: {min_val} - {max_val}")
    else:
        print()

print("\n=== АНАЛИЗ STRING СТОЛБЦОВ ===")
string_columns = df_transactions.select_dtypes(include=['object']).columns
for col in string_columns:
    unique_vals = df_transactions[col].nunique()
    total_vals = len(df_transactions)
    ratio = unique_vals / total_vals
    print(f"{col:25} | Уникальных: {unique_vals:7} | Соотношение: {ratio:.6f}")
    if unique_vals <= 20:  # Покажем все уникальные значения для малых множеств
        print(f"                          Значения: {sorted(df_transactions[col].unique())}")
    print()
```

```python
def optimize_dataframe_memory(df: pd.DataFrame) -> pd.DataFrame:
    """
    Оптимизируем типы данных для экономии памяти.
    
    Args:
        df: DataFrame для оптимизации
        
    Returns:
        pd.DataFrame: Оптимизированный DataFrame
    """
    df_optimized = df.copy()
    
    print("=== ОПТИМИЗАЦИЯ ТИПОВ ДАННЫХ ===")
    
    # 1. Оптимизация integer столбцов (Int64 -> Int32 если возможно)
    int_columns = ['card_number', 'num_transactions', 'unique_merchants', 'unique_countries']
    
    for col in int_columns:
        if col in df_optimized.columns:
            min_val = df_optimized[col].min()
            max_val = df_optimized[col].max()
            
            # Проверяем, помещается ли в int32 (-2,147,483,648 до 2,147,483,647)
            if min_val >= -2**31 and max_val <= 2**31 - 1:
                old_memory = df_optimized[col].memory_usage(deep=True) / 1024**2
                df_optimized[col] = df_optimized[col].astype('int32')
                new_memory = df_optimized[col].memory_usage(deep=True) / 1024**2
                print(f"{col:25} | int64 -> int32 | {old_memory:6.1f} MB -> {new_memory:6.1f} MB")
            else:
                print(f"{col:25} | Остается int64 (значения не помещаются в int32)")
    
    # 2. Оптимизация categorical столбцов
    categorical_candidates = {
        'customer_id': 'Уникальный идентификатор клиента - ограниченное множество',
        'vendor_category': 'Категория продавца - ограниченное множество',
        'vendor_type': 'Тип продавца - ограниченное множество', 
        'vendor': 'Название продавца - ограниченное множество',
        'country': 'Страна - ограниченное множество',
        'city': 'Город - ограниченное множество',
        'city_size': 'Размер города - ограниченное множество',
        'card_type': 'Тип карты - ограниченное множество',
        'device': 'Устройство - может быть много уникальных, но повторяющихся',
        'channel': 'Канал - ограниченное множество',
        'device_fingerprint': 'Уникальный отпечаток устройства - ограниченное множество',
    }
    
    print("\n=== КОНВЕРТАЦИЯ В CATEGORICAL ===")
    for col, description in categorical_candidates.items():
        if col in df_optimized.columns:
            unique_count = df_optimized[col].nunique()
            total_count = len(df_optimized)
            ratio = unique_count / total_count
            
            # Конвертируем в categorical если соотношение уникальных значений < 0.5
            if ratio < 0.5:
                old_memory = df_optimized[col].memory_usage(deep=True) / 1024**2
                df_optimized[col] = df_optimized[col].astype('category')
                new_memory = df_optimized[col].memory_usage(deep=True) / 1024**2
                print(f"{col:25} | string -> category | {old_memory:6.1f} MB -> {new_memory:6.1f} MB | Ratio: {ratio:.4f}")
            else:
                print(f"{col:25} | Остается string (слишком много уникальных: {ratio:.4f})")
    
    return df_optimized

# Применяем оптимизацию
df_transactions_optimized = optimize_dataframe_memory(df_transactions)
df_transactions_optimized
```

```python
# Сравнение результатов оптимизации
print("=== СРАВНЕНИЕ ДО И ПОСЛЕ ОПТИМИЗАЦИИ ===")

# Общее потребление памяти
original_memory = df_transactions.memory_usage(deep=True).sum() / 1024**3
optimized_memory = df_transactions_optimized.memory_usage(deep=True).sum() / 1024**3
savings = original_memory - optimized_memory
savings_percent = (savings / original_memory) * 100

print(f"Исходный размер:      {original_memory:.3f} GB")
print(f"Оптимизированный:     {optimized_memory:.3f} GB")
print(f"Экономия:             {savings:.3f} GB ({savings_percent:.1f}%)")

print("\n=== ДЕТАЛЬНОЕ СРАВНЕНИЕ ТИПОВ ===")
print(f"{'Столбец':25} | {'Было':15} | {'Стало':15} | {'Экономия MB':12}")
print("-" * 80)

for col in df_transactions.columns:
    original_mem = df_transactions[col].memory_usage(deep=True) / 1024**2
    optimized_mem = df_transactions_optimized[col].memory_usage(deep=True) / 1024**2
    savings_col = original_mem - optimized_mem
    
    original_dtype = str(df_transactions[col].dtype)
    optimized_dtype = str(df_transactions_optimized[col].dtype)
    
    if savings_col > 0.1:  # Показываем только значимые изменения
        print(f"{col:25} | {original_dtype:15} | {optimized_dtype:15} | {savings_col:8.1f}")

print(f"\n=== ИНФОРМАЦИЯ ОБ ОПТИМИЗИРОВАННОМ ДАТАФРЕЙМЕ ===")
print(df_transactions_optimized.info(memory_usage='deep'))
```

```python
# Дополнительная проверка: убеждаемся что данные не изменились
print("=== ПРОВЕРКА ЦЕЛОСТНОСТИ ДАННЫХ ===")

# Проверяем, что количество строк осталось прежним
print(f"Исходное количество строк: {len(df_transactions)}")
print(f"После оптимизации: {len(df_transactions_optimized)}")

# Проверяем несколько ключевых столбцов на равенство значений
test_columns = ['transaction_id', 'amount', 'is_fraud', 'ip_address', 'card_number']
for col in test_columns:
    if col in df_transactions.columns:
        are_equal = df_transactions[col].equals(df_transactions_optimized[col])
        print(f"Столбец {col}: {'✓ идентичен' if are_equal else '✗ РАЗЛИЧАЕТСЯ'}")

print("\n=== УНИКАЛЬНЫЕ ЗНАЧЕНИЯ В CATEGORICAL СТОЛБЦАХ ===")
categorical_cols = df_transactions_optimized.select_dtypes(include=['category']).columns
for col in categorical_cols:
    unique_count = df_transactions_optimized[col].nunique()
    categories = sorted(df_transactions_optimized[col].cat.categories.tolist())
    print(f"{col:20} | {unique_count:3} категорий: {categories[:20]}")

# Сохраняем оптимизированный датафрейм
print("\n=== СОХРАНЕНИЕ ОПТИМИЗИРОВАННЫХ ДАННЫХ ===")
output_file = 'transactions_usd_optimized.parquet'
df_transactions_optimized.to_parquet(output_file, compression='snappy')

# Проверяем размер файлов
import os
original_size = os.path.getsize('transactions_usd.parquet') / 1024**2
optimized_size = os.path.getsize(output_file) / 1024**2
file_savings = original_size - optimized_size
file_savings_percent = (file_savings / original_size) * 100

print(f"Размер исходного файла:      {original_size:.1f} MB")
print(f"Размер оптимизированного:    {optimized_size:.1f} MB") 
print(f"Экономия места на диске:     {file_savings:.1f} MB ({file_savings_percent:.1f}%)")
print(f"Файл сохранен как: {output_file}")
```

## Итоговая сводка по оптимизации типов данных

### Применённые оптимизации:

1. **Integer оптимизация (Int64 → Int32)**
   - Экономия: ~50% памяти для этих столбцов (если значения помещаются в int32)

2. **String → Categorical конвертация**
   - Экономия: ~90% памяти для этих столбцов


### Критерии принятия решений:

- **Int32 vs Int64:** Проверяем диапазон значений (-2¹³¹ до 2³¹-1)
- **Categorical vs String:** Соотношение уникальных значений < 0.5 от общего количества
- **Безопасность:** Сохраняем все исходные данные без потерь

### Результаты:
- Экономия памяти: 75% 
- Ускорение операций группировки и фильтрации для categorical столбцов
